```python
import numpy as np
import numpy.linalg as la

x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)

eigvec, _ = la.qr(np.random.randn(2, 2))
eigvals = 0.5 + np.random.rand(2)

covariance_mat = (
    eigvec.dot(np.diag(eigvals).dot(la.inv(eigvec))))

mu = np.random.rand(2) * 0.25
```

```python
def make2DGaussian(x, y, cov, mu):
    inv_cov = np.linalg.inv(cov)
    xi = np.transpose(np.array([x, y]))
    xi_mu = (xi - mu)
    xi_mu_trans = np.transpose(xi_mu)
    mat_mul = xi_mu @ inv_cov @ xi_mu_trans
    return np.exp((-1/2) * mat_mul)
```

```python
z = np.zeros((50, 50))
```

```python
for i in range(50):
    for j in range(50):
        z[i][j] = make2DGaussian(x[i], y[j], covariance_mat, mu)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter


fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(azim=10)
x, y = np.meshgrid(x, y)
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0.1, antialiased=True)
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def getNpeople(n):
    return np.random.randint(0, 365, n)

def hasDups(birthdays):
    return len(birthdays) != len(set(birthdays))

def getProb_n(n):
    if n == 1:
        return 0
    counter = 0
    for i in range(50000):
        birthdays = getNpeople(n)
        if hasDups(birthdays):
            counter += 1
    return counter / 50000

def getPerc_50(prob_n):
    for i in range(len(prob_n)):
        if prob_n[i] > 0.5:
            return i + 2
    else:
        return -1

prob_n = np.array([getProb_n(n) for n in range(2, 101)])
perc_50 = getPerc_50(prob_n)

n = np.linspace(2, 100, 99, dtype='I')
plt.plot(n, prob_n)
plt.xlabel('n')
plt.ylabel('prob_n')
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

n_term_max = 20
def taylor(k,a,x):
    terms = np.exp(np.ones(k) * a)
    for i in range(len(terms)):
        terms[i] *= ((x - a)**i / (math.factorial(i)))
    return sum(terms)

# define X range
X = np.linspace(-4.,4.,160)

# set X and Y axis of plot
axes = plt.gca()
axes.set_xlim([-4.,4.])
axes.set_ylim([-5.,50.])

#define taylor_vals and error_vals
n = len(X)
taylor_vals = np.array([[taylor(i, 1, x) for x in X] for i in range(n_term_max)])
error_vals = np.abs(taylor_vals - [[np.exp(x) for x in X] for i in range(n_term_max)])

# Here add code to compute taylor_vals and error_vals

#plotting code
plt.plot(X,[np.exp(x) for x in X],'bo',label = 'true value')
for i in range(18, n_term_max):
    plt.plot(X, taylor_vals[i], label = 'Order ' + str(i))
plt.legend()
# create plots of Taylor expansions here

#plots for errors come here
plt.figure()
for i in range(18, n_term_max):
    plt.semilogy(X, error_vals[i], label = 'Order ' + str(i))
plt.legend()
plt.show()
```

```python
import sympy as sp
```

```python
x = sp.Symbol("x")
```

```python
from math import factorial
```

```python
def C(f, n):
    return abs(f.diff(x, n + 1)) / factorial(n + 1)
```

```python
f1 = -sp.sin(x) - sp.sin(2 * x) / 8
```

```python
f2 = (x - 1) / (x + 1)
```

```python
C1 = C(f1, 2)
```

```python
C2 = C(f2, 2)
```
